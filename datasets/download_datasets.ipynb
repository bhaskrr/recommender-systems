# Dataset Downloader

Run any cell below to download the dataset needed. Files are extracted automatically into this folder.

| Name           | Used In               | Size |
| -------------- | --------------------- | ---- |
| MovieLens 100K | Foundational Paradigm | ~1MB |


```python
import urllib.request
import zipfile
import os
from pathlib import Path

DATASETS_DIR = Path(".") # Current folder(datasets/)

DATASETS = {
    "ml-100k": {
        "url": "https://files.grouplens.org/datasets/movielens/ml-latest-small.zip",
        "filename": "ml-100k.zip",
        "format": "zip",
    }
}

def download_dataset(name: str):
    """Download and extract a dataset by name."""
    
    if name not in DATASETS:
        print(f"Unknown dataset '{name}'. Available: {list(DATASETS.keys())}")
        return
    
    dataset = DATASETS[name]
    dest_dir = DATASETS_DIR/name
    archive_path = DATASETS_DIR/dataset["filename"]
    
    if dest_dir.exists():
        print(f"'{name}' already exists at {dest_dir}. Skipping download.")
        return
    
    print(f"Downloading {dataset['filename']}...")
    
    def progress(block, block_size, total):
        downloaded = block * block_size
        if total > 0:
            pct = min(downloaded / total * 100, 100)
            print(f"\r  {pct:.1f}%", end="", flush=True)
    
    urllib.request.urlretrieve(dataset["url"], archive_path, reporthook=progress)
    
    print(f"\n Extracting...")
    
    if dataset["format"] == "zip":
        with zipfile.ZipFile(archive_path, "r") as z:
            z.extractall(DATASETS_DIR)
    
    archive_path.unlink()
    
    print(f"  Done. Saved to {dest_dir}")
    
print("Downloader ready. Available datasets:", list(DATASETS.keys()))
```

## Download MovieLens 100K

```python
download_dataset("ml-100k")
```
